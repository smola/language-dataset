# A short document on the image analysis  and statistics for multiphoton FRET 

Using two-photon techniques for performing FRET-based stoichiometry requires including additional terms as compared to the one photon case. Namely, we cannot ignore the excitation of donor due to the acceptor excitation as well as the emission of acceptor into the donor emission channel. This has been detailed wonderfully in Meredith's thesis and I will not repeat it here. Instead, I will outline the image processing techniques that are used in the data analysis developed by Dan and Amar. The basic algorithm is as follows:

* Use Donor only and FRET only cells
    - measure lifetimes in SMART center. Get an average value of the FRET efficiency, E

* Donor only cells
    - pick out individual cells from each image, average $I_d$, $I_n$, $I_f$ over each cell
    - calculate $\beta$ and $\theta$
    - repeat for each image
    - calculate average $\beta$ and $\theta$
    - 74 cells used for beta
    - 66 cells used for theta
    
    Shown below is an example of how individual cells are thresholded (on the right) against the background and then each cell is individually selected (on the left, shown by the color coding) and calculations are performed on the signal from the entire cell
    
    ![example of picking cells for $\beta$ calculation](alpha_images/cell_pick_0.png)
    
    
* Acceptor only cells
    - pick out individual cells from each image, average $I_a$, $I_n$, $I_f$ over each cell
    - calculate $\alpha$ and $\eta$
    - repeat for each image
    - calculate average $\alpha$ and $\eta$
    - 28 cells used for $\alpha$
    - 29 cells used for $\eta$
    
    
* Use FRET only cells
    - pick out individual cells, average $I_a$, $I_d$, $I_f$ over each cell
    - use the average values of E, $\beta$, $\theta$, $\alpha$ and $\eta$ as calculated above
    - calculate using \begin{equation} \gamma = \frac{E}{-1 + \frac{\eta \left(- \alpha + \theta\right) \left(- I_{d} \beta + I_{f}\right)}{\alpha \left(- \beta + \eta\right) \left(I_{a} \theta - I_{f}\right)}} \end{equation}
    - repeat for each image
    - calculate average $\gamma$
    - 77 cells used for calculating $\gamma$
    
    In the graph below, the histogram of $\gamma$ is shown.
    
    ![distribution of $\gamma$](alpha_images/gamma_hist.png)
   
   
* Use FRET only cells
    - pick out individual cells, average $I_a$, $I_d$, $I_f$ over each cell
    - use the average values of E, $\beta$, $\theta$, $\alpha$, $\eta$ and $\gamma$ as calculated above
    - calculate using \begin{equation} \xi = \frac{E \gamma \left(I_{d} \eta - I_{f}\right)}{\left(- E + 1\right) \left(- \beta + \eta\right) \left(- \frac{I_{a} \alpha \theta}{- \alpha + \theta} - \frac{I_{d} \beta \eta}{- \beta + \eta} + \frac{I_{f} \left(- \alpha \beta + \eta \theta\right)}{\left(- \alpha + \theta\right) \left(- \beta + \eta\right)}\right)} \end{equation}
    - repeat for each image
    - calculate average $\xi$ 
    - also 77 cells, because the same data set is used as $\gamma$
    
    In the graph below, the histogram of $\xi$ is shown.
    
    ![distribution of $\gamma$](alpha_images/xi_hist.png)
    
    
* Use FRET only cells (trivial case sanity check)
    - Individual cells have already been picked out above for $\gamma$ and $\xi$ calculation. Use the averaged $I_a$, $I_d$, $I_f$ over each cell as above. 
    - use all the above quantities to calculate fraction of donor $f_d$, and acceptor $f_a$
    - Use \begin{equation} 
    f_{a} = \frac{\gamma}{E} \left(-1 + \frac{\eta \left(- \alpha + \theta\right) \left(- I_{d} \beta + I_{f}\right)}{\alpha \left(- \beta + \eta\right) \left(I_{a} \theta - I_{f}\right)}\right) 
    \end{equation}
    - Use \begin{equation} f_{d} = \frac{1}{E} \left(1 - \frac{I_{d} \eta - I_{f}}{\left(- \beta + \eta\right) \left(\frac{I_{d} \eta - I_{f}}{- \beta + \eta} + \frac{\xi}{\gamma} \left(- \frac{I_{a} \alpha \theta}{- \alpha + \theta} - \frac{I_{d} \beta \eta}{- \beta + \eta} + \frac{I_{f} \left(- \alpha \beta + \eta \theta\right)}{\left(- \alpha + \theta\right) \left(- \beta + \eta\right)}\right)\right)}\right) \end{equation} 
    - confirm that they average to one (trivial case because $\gamma$ and $\xi$ are calculated assuming $f_a$ and $f_d$ are one for FRET case samples 
    
    Shown below are the histograms of $f_a$ and $f_d$ for the FRET only case. The histograms should be centered about 1. Also included are the same numbers plotted $f_a$ vs $f_d$. Ideally, this distribution should be a Gaussian distribution around the point (1,1) in this image.
    
    ![distribution of $f_a$, $f_d$ and $f_a$ vs $f_d$ for FRET](alpha_images/favsfd_FRET.png)
    
    
* Use unlinked donor and acceptor cells (sanity check)
    - pick out individual cells, average $I_a$, $I_d$, $I_f$ over each cell
    - use all the above quantities to calculate fraction of donor $f_d$, and acceptor $f_a$
    - confirm that they average to zero
    - Use same formulas as above for $f_a$ and $f_d$ 
    - 45 cells used for calculating $f_a$ and $f_d$
    
    Shown below are the histograms of $f_a$ and $f_d$ for the donor and acceptor unlinked only case. The histograms should be centered about 0. Also included are the same numbers plotted $f_a$ vs $f_d$. Ideally, this distribution should be a Gaussian distribution around the point (0,0) in this image.
    
    ![distribution of $f_a$, $f_d$ and $f_a$ vs $f_d$ for no FRET](alpha_images/favsfd_noFRET.png)
    
    
Here lies the worrying part. The distribution of $f_a$ and $f_d$ should not display significant covariance with respect to each other, since they should be random quantities. Is there an underlying relation that causes those numbers to naturally vary together because many of the numbers used for their calculations are the same? Would a completely random set of $I_a$, $I_d$ and $I_f$ still generate this covariance?
